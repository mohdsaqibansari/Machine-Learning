```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/Acer/Desktop/KNN/Social_Network_Ads.csv")
data.head()
```

```python
X=data.iloc[:,2:4]
X.values
```

```python
y=data.iloc[:,-1]
y.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scale=scaler.fit_transform(X_train)
X_train_scale
```

```python
X_test_scale=scaler.transform(X_test)
X_test_scale
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# First Method
```

```python
np.sqrt(X_train_scale.shape[0])
```

```python
k=7
```

```python
my_knc=KNeighborsClassifier(n_neighbors=17)
my_knc.fit(X_train_scale,y_train)
```

```python
my_first_pred=my_knc.predict(X_test_scale)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
my_first_acc=accuracy_score(my_first_pred,y_test)
my_first_acc
```

```python
# Second Method
```

```python
accuracy=[]
for i in range(1,26):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scale,y_train)
    accuracy.append(accuracy_score(y_test,knn.predict(X_test_scale)))
accuracy
```

```python
plt.plot(range(1,26,1),accuracy)
```

```python
max(accuracy)
```

```python
for i in range(25):
    print(accuracy[i], " :" , i+1 )
```

```python
my_new_knn=KNeighborsClassifier(n_neighbors=9)
my_new_knn.fit(X_train_scale,y_train)
```

```python
final_pred=my_new_knn.predict(X_test_scale)
```

```python
accuracy_score(final_pred,y_test)
```
